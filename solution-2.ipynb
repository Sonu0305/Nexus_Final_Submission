```python
main_vars14 = ['ID', 'DRUGNAME', 'GENENAME', 'Disease']
X14_train = df[main_vars14]
y14_train = df['Target_Status']

X14_test = df2[main_vars14]

X14_train_encoded = pd.get_dummies(X14_train.drop(columns=['ID']), drop_first=True)
X14_test_encoded = pd.get_dummies(X14_test.drop(columns=['ID']), drop_first=True)

X14_test_encoded = X14_test_encoded.reindex(columns=X14_train_encoded.columns, fill_value=0)

scaler14 = StandardScaler()
X14_train_scaled = scaler14.fit_transform(X14_train_encoded)
X14_test_scaled = scaler14.transform(X14_test_encoded)

model14 = svm.SVC(kernel='linear', random_state=42)  
model14.fit(X14_train_scaled, y14_train)

y14_pred = model14.predict(X14_test_scaled)

submission14 = pd.DataFrame({
    'ID': df2['ID'], 
    'Prediction': y14_pred
})
submission14.to_csv('submission14.csv', index=False)
```
