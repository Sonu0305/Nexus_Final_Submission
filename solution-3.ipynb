```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

main_vars15 = ['ID', 'DRUGNAME', 'GENENAME', 'SEQUENCE', 'Disease']
X15_train = df[main_vars15].copy()  # Create a copy of the DataFrame
y15_train = df['Target_Status']

X15_test = df2[main_vars15].copy()  # Create a copy of the DataFrame

# Apply pd.factorize for encoding
for col in main_vars15:
    if col != 'ID':
        # Use .loc to avoid SettingWithCopyWarning
        X15_train.loc[:, col], _ = pd.factorize(X15_train[col])
        X15_test.loc[:, col], _ = pd.factorize(X15_test[col])

X15_train_encoded = X15_train.drop(columns=['ID'])
X15_test_encoded = X15_test.drop(columns=['ID'])

model15 = RandomForestClassifier(random_state=42)
model15.fit(X15_train_encoded, y15_train)

y15_pred = model15.predict(X15_test_encoded)

submission15 = pd.DataFrame({
    'ID': df2['ID'], 
    'Prediction': y15_pred
})

submission15.to_csv('submission15.csv', index=False)
```
